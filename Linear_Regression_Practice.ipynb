```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
# Load the dataset
diabetes = load_diabetes()
diabetes_df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
diabetes_df['TARGET'] = diabetes.target
diabetes_df
```

```python
# Split the dataset into training and testing sets
X = diabetes_df.drop(columns="TARGET")
y = diabetes_df['TARGET']
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Fit the linear regression model
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
# Interpret the regression information
# Regression Coefficients
coefficients = pd.DataFrame({'Feature': X.columns, 'Coefficient': lm.coef_})
print("Regression Coefficients:")
print(coefficients)
```

```python
# Regression P-Value
X_train_with_const = sm.add_constant(X_train)
model = sm.OLS(y_train, X_train_with_const).fit()
p_values = model.pvalues[1:]  # Exclude the constant term
print("\nRegression P-Values:")
print(p_values)
```

```python
# Regression R-Squared
print("\nRegression R-Squared:", lm.score(X_train, y_train))
```
